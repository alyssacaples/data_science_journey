```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('transactions.csv')
#dropping columns
df.head(5)
```

```python
df = df.drop(columns=['ID', 'Type', 'Unnamed: 7', 'Unnamed: 8', 'Unnamed: 9', 'Unnamed: 10', 'Unnamed: 11', 'Unnamed: 5', 'Unnamed: 6'])
```

### Format Data

```python
df['Date'] = df['Date'].str[:-4]
df
```

```python
df['Amount'] = pd.to_numeric(df['Amount'])
df
```

```python
import re

df['Name'] = df['Name'].apply(lambda x: re.sub(r'\b\d+\b', "", x))

index_names = df[ df['Name'].str.contains("VENMO")].index 
df.drop(index_names, inplace = True)

df['Name'] = df['Name'].apply(lambda x: x.replace("SQ ", ""))
df['Name'] = df['Name'].apply(lambda x: x.replace("SP ", ""))
df['Name'] = df['Name'].apply(lambda x: x.replace("*", ""))
df['Name'] = df['Name'].apply(lambda x: x.replace("//", ""))
df['Name'] = df['Name'].apply(lambda x: x.replace("#", ""))
df['Name'] = df['Name'].apply(lambda x: x.replace("-", ""))



#text_after = re.sub(regex_search_term, regex_replacement, text_before)


df.head(30)

  
# drop these row indexes 
# from dataFrame 
```

```python
#use .apply()
df['City'] = df['Name'].apply(lambda x: x.split()[-2:-1][0].upper())
df['State'] = df['Name'].apply(lambda x: x.split()[-1])
df['Name'] = df['Name'].apply(lambda x: x.replace("FL", ""))
df['Name'] = df['Name'].apply(lambda x: x.replace("Gainesville", ""))
df['Name'] = df['Name'].apply(lambda x: x.replace("GAINESVILLE", ""))
df['Name'] = df['Name'].apply(lambda x: x.replace("OVIEDO", ""))
df['Name'] = df['Name'].apply(lambda x: x.replace("Oviedo", ""))
df['Name'] = df['Name'].apply(lambda x: x.replace("ORLANDO", ""))
```

```python
index_names = df[ df['State'] != "FL"].index 
df.drop(index_names, inplace = True)
df
```

```python
index_names = df[ df['Amount'] > 0].index 
df.drop(index_names, inplace = True)
df
```

```python
valid_cities = ['OVIEDO', 'GAINESVILLE', 'LARGO', 'PETERSBURG', 'ORLANDO' ]

index_names = df[ ~df['City'].isin(valid_cities) ].index 
df.drop(index_names, inplace = True)
df.loc[14, 'Name'] ="MCDONALD'S"
df.loc[28, 'Name'] ="MICHAEL'S"
df.loc[41, 'Name'] ="MICHAEL'S"
df.loc[56, 'Name'] = "BLACK CROW COFFEE CO"
df
```

```python
df.loc[(df['Name'].str.contains("COFFEE")),'Name']='COFFEE'
df.loc[(df['Name'].str.contains("BAGEL")),'Name']='BAGEL'
df
```

```python
df['Amount'] = df['Amount']*-1
df
```

```python
df_sort = df.sort_values(by=['Amount'], ascending=False)
top_ten = df_sort[:10]
top_ten
```

```python
money = top_ten['Amount'].astype(str)

#explode = (0, 0.1, 0, 0)  # only "explode" the 2nd slice (i.e. 'Hogs')
labels = top_ten['Name'].astype(str)


#.astype(str)


fig1, ax1 = plt.subplots()
ax1.pie(money, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

# plt.show()
```
