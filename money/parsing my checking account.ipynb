```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('transactions.csv')

#dropping columns
df = df.drop(columns=['Account', 'ID', 'Type',  'Balance', 'Unnamed: 7', 'Unnamed: 8', 'Unnamed: 9', 'Unnamed: 10', 'Unnamed: 11', 'Unnamed: 12', 'Unnamed: 6'])
df.head(5)
```

### Format Data

```python
df['Date'] = df['Date'].str[:-4]
df
```

```python
df['Amount'] = pd.to_numeric(df['Amount'])
df
```

```python


#use .apply()
df['Location'] = df['Name'].apply(lambda x: x.split()[-2:-1][0])
df['Store'] = df['Name'].apply(lambda x: x.split()[:-2])

df['Store'] = df['Store'].apply(lambda x: x[0])

df
```
