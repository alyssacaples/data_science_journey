```python
import requests
from bs4 import BeautifulSoup as bs
import re
```

```python
#get info from webpage
r = requests.get("https://everydaypower.com/jaden-smith-quotes/")

#convert to a beautiful soup object
soup = bs(r.content)

#prints out elements in a pretty way
print(soup.prettify()) 
```

```python
headers = soup.find_all("p")
quotes = []
for s in headers:
    #print(s.string)
    try:
        print(s.string)
        quotes.append(re.match(s.get_text(), r"(<p>)\d+\.").string)
    except Exception as e:
        print(s.get_text())
        print(e)

print("quotes:")
print(quotes)
```
